```python
from __future__ import annotations
from typing import TYPE_CHECKING, Any

if TYPE_CHECKING:
    from typing_extensions import Self
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
import numpy.typing as npt


class PerformancePlot:
    def __init__(self, dpi: int, figsize: tuple[int, int]) -> None:
        self.fig = plt.figure(figsize=figsize, dpi=dpi)
        self.ax = plt.subplot()

    def plot(
        self,
        color: str,
        x_label: str,
        y_label: str,
        x_ticks: list[str],
        Y: npt.NDArray[np.float64 | np.int64],
        X: npt.NDArray[np.float64 | np.int64] | None = None,
        y_max: int = 500,
    ) -> Self:
        X = X or np.arange(len(Y))

        rects = self.ax.bar(X, Y, width=0.3, color=color)
        self.ax.bar_label(rects, padding=3)

        self.ax.xaxis.set_ticks(X.tolist(), x_ticks, fontsize=12, family="serif")
        self.ax.set_ylabel(y_label, labelpad=10.0, fontsize=12, family="serif")
        self.ax.set_xlabel(x_label, labelpad=10.0, fontsize=12, family="serif")

        self.ax.grid(alpha=0.5, axis="y")
        self.fig.tight_layout()
        self.fig.subplots_adjust(right=0.92)

        self.ax.set_ylim(0, y_max)

        return self
```

```python
class MatrixPlot:
    def __init__(self, dpi: int, figsize: tuple[int, int]) -> None:
        self.fig = plt.figure(figsize=figsize, dpi=dpi)
        self.ax = plt.subplot()

    def plot(
        self,
        color: str,
        x_label: str,
        y_label: str,
        x_ticks: list[str],
        Y: npt.NDArray[np.float64 | np.int64],
        X: npt.NDArray[np.float64 | np.int64] | None = None,
        y_max: int = 500,
    ) -> Self:
        X = X or np.arange(len(Y))

        rects = self.ax.bar(X, Y, width=0.3, color=color)
        self.ax.bar_label(rects, padding=3)

        self.ax.xaxis.set_ticks(X.tolist(), x_ticks, fontsize=12, family="serif")
        self.ax.set_ylabel(y_label, labelpad=10.0, fontsize=12, family="serif")
        self.ax.set_xlabel(x_label, labelpad=10.0, fontsize=12, family="serif")

        self.ax.grid(alpha=0.5, axis="y")
        self.fig.tight_layout()
        self.fig.subplots_adjust(right=0.92)

        self.ax.set_ylim(0, y_max)

        return self


class Column:
    def __init__(self, name: str, color: str, value: float) -> None:
        self.name = name
        self.color = color
        self.value = value
```
